```python
from PIL import Image
```

```python
from torchvision import transforms as T
```

```python
img = Image.open("/content/dog.7809.jpg")
```

```python
transform = T.Compose([
    T.Resize((224,224)),
    T.RandomRotation((30, 45))
])
```

```python

transform(img).show()
```

```python
transform = T.Compose([

    T.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1)
])
colorshifting_img = transform(img)
colorshifting_img.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Load the two images
# Create a new figure and subplot with 1 row and 2 columns
fig, axes = plt.subplots(1, 2)

# Display the first image on the left subplot
axes[0].imshow(img)
axes[0].axis('off')

# Display the second image on the right subplot
axes[1].imshow(colorshifting_img)
axes[1].axis('off')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.1)

# Show the final composite image
plt.show()
```

```python
transform = T.Compose([
    T.RandomHorizontalFlip(0.9)
])
flip_img= transform(img)
flip_img.show()
```

```python
fig, axes = plt.subplots(1, 2)

# Display the first image on the left subplot
axes[0].imshow(img)
axes[0].axis('off')

# Display the second image on the right subplot
axes[1].imshow(flip_img)
axes[1].axis('off')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.1)

# Show the final composite image
plt.show()
```

```python
zoom_factor = 0.6 # Change this value to control the zoom level
original = (img.height, img.width)
crop_size = (img.height*zoom_factor, img.width *zoom_factor)  # Keep the original size

# Create a transform composition
transform = T.Compose([
    T.CenterCrop(crop_size),  # Center crop to original size
    T.Resize(original),  # Resize to zoom in
])

# Apply the transform to the image
zoomed_image = transform(img)

# Display the zoomed image
zoomed_image.show()
```

```python
fig, axes = plt.subplots(1, 2)

# Display the first image on the left subplot
axes[0].imshow(img)
axes[0].axis('off')

# Display the second image on the right subplot
axes[1].imshow(zoomed_image)
axes[1].axis('off')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.1)

# Show the final composite image
plt.show()
```

```python
transform = T.Compose([
    T.RandomPerspective(distortion_scale=0.5, p=1, interpolation=Image.BILINEAR)
])
perspective_img = transform(img)
perspective_img.show()
```

```python
fig, axes = plt.subplots(1, 2)

# Display the first image on the left subplot
axes[0].imshow(img)
axes[0].axis('off')

# Display the second image on the right subplot
axes[1].imshow(perspective_img)
axes[1].axis('off')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.1)

# Show the final composite image
plt.show()
```
